```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```
sns.load_dataset
```

```
exercise =sns.load_dataset('exercise')
```

```
exercise
```

# Univariate → “one variable” data visualization

```
#strip plot

sns.stripplot(y= 'diet', data=exercise,jitter=True)
plt.ylabel('diet ($)')
plt.show()
```

```
#Grouping with stripplot()
sns.stripplot(x='pulse', y='diet', data=exercise)
plt.ylabel('diet ($)')
plt.show()
```

```
sns.stripplot(x='pulse', y='diet', data=exercise, size=3,jitter=True)
plt.ylabel('diet ($)')
plt.show()
```

```
#Swarm plot

sns.swarmplot(x='pulse', y='diet', data=exercise)
plt.ylabel('diet ($)')
plt.show()
```

```
sns.swarmplot(x='pulse', y='diet', data=exercise, hue='id')
plt.ylabel('diet ($)')
plt.show()
```

```
#Changing orientation
sns.swarmplot(x='pulse', y='diet', data=exercise, hue='id',orient='h')
plt.xlabel('pulse ($)')
plt.show()
```

# Box and Violin plot

```
#plt.subplot(1,2,1)
sns.boxplot(x='pulse', y='diet', data=exercise)
plt.ylabel('diet ($)')
```

```
#plt.subplot(1,2,2)
sns.violinplot(x='pulse', y='diet', data=exercise)
plt.ylabel('diet ($)')

plt.show()
```

```
plt.subplot(1,2,1)
sns.boxplot(x='pulse', y='diet', data=exercise)
plt.ylabel('diet ($)')
plt.subplot(1,2,2)
sns.violinplot(x='pulse', y='diet', data=exercise)
plt.ylabel('diet ($)')

plt.show()
```

```
##Combining plots

sns.violinplot(x='pulse', y='diet', data=exercise, inner=None,color='lightblue')
sns.stripplot(x='pulse', y='diet', data=exercise, size=3,jitter=True)
plt.ylabel('exercise ($)')
plt.show()
```

# Bivariate → “two variables” data visualization

```
#Joint plot

sns.jointplot(x= 'kind', y= 'diet', data=exercise)
plt.show()
```

```
# Density plot
sns.jointplot(x='id', y= 'pulse', data=exercise,kind='kde')
plt.show()
```

```
#Pair plot

sns.pairplot(exercise)
plt.show()
```

```
sns.pairplot(exercise, hue='id')
plt.show()
```
