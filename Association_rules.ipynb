# ***Book***

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
book = pd.read_csv("/content/book.csv")
book.head()
```

```python
book=pd.get_dummies(book)
book.head()
```

```python
frequent_itemsets = apriori(book, min_support=0.0, use_colnames=True)
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=0.3)
rules
```

```python
rules.sort_values('lift',ascending = False)[0:10]
```

```python
rules[rules.lift>1]
```

# ***Movies***

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
movies = pd.read_csv("/content/my_movies.csv")
movies.head()
```

```python
df=pd.get_dummies(movies)
df.head()
```

```python
frequent_itemsets = apriori(df, min_support=0.2, use_colnames=False)
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=0.8)
rules
```
